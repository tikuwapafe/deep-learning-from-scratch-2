#### 1.2 ニューラルネットワークの推論

```python
# ミニバッチ版の全結合層による変換
W1 = np.random.randn(2,4)  # 重み
b1 = np.random.randn(4)    # バイアス
x = np.random.randn(10,2)  # 入力
h = np.dot(x,W1) + b1      # 隠れ層のニューロン
```

```python
# シグモイド関数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# 隠れ層のニューロンを変換
a = sigmoid(h)
```

```python
# aをまた別の全結合層によって変換
W2 = np.random.randn(4,3)  
b2 = np.random.randn(3)  
s = np.dot(a,W2) + b2
```
